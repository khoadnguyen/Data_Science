## Random Walk - Function Approach

```python
def roll_dice():
    return np.random.randint(1,7)
```

```python
def dice_logic(step, dice):
    
    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + roll_dice()
        
    # Implement clumsiness
    if np.random.rand() <= 0.001 :
        step = 0
        
    return step
```

```python
def dice_game(init_step):
    dice = roll_dice()
    next_step = dice_logic(init_step, dice)
    return next_step
```

```python
def sim_rand_walk(steps):
    
    random_walk = [0]
    for x in range(steps):
        step = random_walk[-1]
        next_step = dice_game(step)
        random_walk.append((next_step))
        
    return random_walk
```

```python
def run_sims(num_sims):
    
    all_walks = []
    for x in range(num_sims):
        rand_walk = sim_rand_walk(100)
        all_walks.append(rand_walk)
        
    return np.transpose(all_walks)
```

```python
def plot_walks(all_walks):
    plt.plot(all_walks)
    plt.title("All Walks")
    plt.ylabel("Steps")
    plt.xlabel("Dices Rolled")
    plt.show()
```

```python
def plot_distribution(final_steps):
    plt.hist(final_steps, bins=20)
    plt.title("Random Walk Outcome Distribution")
    plt.ylabel("Num Simulations")
    plt.xlabel("Steps Reached")
    plt.show()
```

```python
num_sims = 100
all_walks = run_sims(num_sims)

plot_walks(all_walks)
plot_distribution(all_walks[-1])
```

```python
final_steps = all_walks[-1]
fav_outcomes = final_steps[np.where(final_steps > 60)]

print(len(fav_outcomes)/num_sims)
```
