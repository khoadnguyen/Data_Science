## Load Seasons Data

```python
nba = pd.read_csv("../../data/nba_1950/season_stats.csv")
```

### Use `value_counts()` to view all unique values in a distribution and their frequencies.

```python
shoot = nba['TS%'].value_counts()
shoot.head()
```

```python
sorted_shoot = nba['TS%'].value_counts().sort_index()
sorted_shoot.head()
```

### Use matplotlib to plot a Histogram to Visualize a Distribution

```python
_ = plt.hist(nba['TS%'], bins=20)

# Make it pretty 
_ = plt.title("True Shooting % Across all NBA Seasons")
_ = plt.xlabel("True Shooting Percentage")
_ = plt.ylabel("# of Players")

# "print" the plot
plt.show()
```

### How different is the NBA Now than in 1950?

```python
first_year = nba['Year'].min()
last_year = nba['Year'].max()
print(first_year, last_year)
```

```python
y_1950 = nba[nba['Year'] == first_year]
```

```python
y_2017 = nba[nba['Year'] == last_year]
```

```python
_ = plt.hist(y_1950['TS%'], bins=20)

# Make it pretty 
_ = plt.title("True Shooting % for the 1950 Season")
_ = plt.xlabel("True Shooting Percentage")
_ = plt.ylabel("# of Players")

# "print" the plot
plt.show()
```

```python
_ = plt.hist(y_2017['TS%'], bins=20)

# Make it pretty 
_ = plt.title("True Shooting % for 2017 Season")
_ = plt.xlabel("True Shooting Percentage")
_ = plt.ylabel("# of Players")

# "print" the plot
plt.show()
```

## Load Player Data

```python
players = pd.read_csv('../data/nba_1950/player_data.csv')
```

### Does The College Attended affect performance in NBA?

```python
top_5_colleges = ['University of Kentucky',
                  'Duke Univeristy',
                  'University of Kansas',
                  'Syracuse University',
                  'University of California, Los Angeles',
                 ]
```

```python
# Check "college" column against top_5_colleges
in_top_5_colleges = players['college'].isin(top_5_colleges)

# print(in_top_5_colleges)
```

```python
# index entire players df by "in_top_5"
players_in_top_5 = players[in_top_5_colleges]['name']
```

```python
top_college = nba[nba['Player'].isin(players_in_top_5)]
top_college.shape
```

```python
bottom_college = nba[~nba['Player'].isin(players_in_top_5)]
bottom_college.shape
```

#### Plot the Players not from Top 5

```python
_ = plt.hist(bottom_college['PER'], bins=50)

# Make it pretty 
_ = plt.title("PER for Players NOT from Top 5 Colleges")
_ = plt.xlabel("Player Efficiency Rating")
_ = plt.ylabel("# of Players")

# "print" the plot
plt.show()
```

#### Plot the Players from Top 5

```python
_ = plt.hist(top_college['PER'], bins=50)

# Make it pretty 
_ = plt.title("PER for Players FROM Top 5 Colleges")
_ = plt.xlabel("Player Efficiency Rating")
_ = plt.ylabel("# of Players")

# "print" the plot
plt.show()
```

What conclusions can we draw about these 2 groups?

```python
# There is no significant difference in NBA performance for players 
# that were drafted entirely out of the top 5 colleges and those who weren't.
```
